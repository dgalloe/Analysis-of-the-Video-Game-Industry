## Preparar los datos

```python
# Información general del Dataframe
video_games.info(show_counts=True)
```

```python
# Revisión estructura Dataframe
print(video_games.sample(10))
```

```python
# Reemplazar nombres de columnas a minuscula
video_games.columns = video_games.columns.str.lower()
video_games.columns
```

```python
# Reemplazar valores 'NaN' en columma 'year_of_release' por cero. 
video_games['year_of_release'].fillna(0, inplace=True)

# Convertir la columna 'year_of_release' en 'intenger'
video_games['year_of_release'] = video_games['year_of_release'].astype('int')
#print(video_games['year_of_release'].unique())

# Reemplazar valores 'NaN' en columma 'name' and genre' por no definido. 
video_games['name'].fillna('undefined', inplace=True)
video_games['genre'].fillna('undefined', inplace=True)
#null_filter = video_games[video_games['genre'].isna()]
#print(video_games['genre'].unique())
#print(null_filter)

# Reemplazar valores 'NaN' de las colummas 'critic_score' and 'user_score' por cero, and 'rating' por no definido. 
video_games['critic_score'].fillna(0, inplace=True)
video_games['user_score'].fillna(0, inplace=True)
video_games['rating'].fillna('undefined', inplace=True)

# Convertir la columna 'critic_score' en 'intenger'
video_games['critic_score'] = video_games['critic_score'].astype('int')
```

```python
# Revisar datos únicos de las columnas 'critic_score', 'user_score' and 'rating'. 

print(video_games['critic_score'].unique())
print()
print(video_games['user_score'].unique())
print()
print(video_games['rating'].unique())
```

```python
# Replicar la columna 'user_score' solo con datos númericos, 'tbd' será igual a cero. 
video_games['user_score_numeric'] = pd.to_numeric(video_games['user_score'], errors='coerce')
video_games['user_score_numeric'].fillna(0, inplace=True)
print(video_games.sample(10))
```

```python
# Revisar nuevamente datos generales del Dataframe
video_games.info()
```

```python
# Calcular ventas totales para cada juego

#print(video_games['name'].nunique())
#conteo_juegos = video_games['name'].value_counts()
#print(conteo_juegos)
#print(video_games[video_games['name']== 'Ratatouille'])

video_games['total_sales'] = video_games['na_sales'] +video_games['eu_sales']+video_games['jp_sales'] +video_games['other_sales']
print(video_games.sample(10))
```

Valores ausentes: 
- Se reemplazaron los valores ausentes de las columnas 'year_of_release', 'critic_score' y 'user_score' por cero. Se define tratarlos valores ausentes de esta forma, ya que posteriormente es probable que tengamos que realizar calculos con estas columnas, por lo tanto, los valores ausentes deben tomar un valor numérico. 
- Se reemplazaron los valores ausentes de las columnas 'name', 'genre' y 'rating' por no definifo. Estas columnas son de texto, pero lo tanto, para poder trabajar con ellas se realiza este cambio. 
- Para la columna 'rating' que contiene el valor 'tbd' y que se debe mantener, se define replicar esta columna solo con valores numericos, para poder realizar cálculos más adelante. 'tbd' toma el valor de cero.

```python
#Elegir las plataformas con las mayores ventas totales y construir una distribución basada en los datos de cada año. 
top_sales_platforms = ['PS2','X360','PS3','Wii','DS','PS']
most_sold_platforms = video_games.query("platform in @top_sales_platforms").copy()
most_sold_platforms = most_sold_platforms[most_sold_platforms['year_of_release'] > 0]
most_sold_platforms = most_sold_platforms.groupby(['year_of_release','platform'])['total_sales'].sum().reset_index()
print(most_sold_platforms)

print()

plt.figure(figsize=(12, 6))
hist_plot = sns.histplot(
    data=most_sold_platforms,
    x='year_of_release',
    weights='total_sales',  
    bins=30,                
)


plt.title('Sales distribution by year (Most sold platforms)')
plt.xlabel('Year of Release')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)     
plt.tight_layout()          
plt.xlim(left=1992)
plt.show()
```

```python
# Buscar las plataformas que solían ser populares pero que ahora no tienen ventas.

# Calcular el último año con ventas para cada plataforma
last_active_year = video_games.groupby('platform')['year_of_release'].max()

# Obtener el año más reciente en el dataset
current_year = video_games['year_of_release'].max()

# Filtrar plataformas inactivas (sin ventas en los últimos 5 años)
inactive_threshold = 5  
obsolete_platforms = last_active_year[last_active_year <= (current_year - inactive_threshold)]
#print(obsolete_platforms)

# Combinar para traer las ventas totales
obsolete_platforms_cumul_sales = pd.merge(obsolete_platforms,video_games_by_platform_order,on=['platform'], how= 'left')
obsolete_platforms_cumul_sales = obsolete_platforms_cumul_sales.sort_values(by='total_sales', ascending=False)
obsolete_platforms_cumul_sales.columns = ['platform','last_active_year','total_sales']

# Top 5 de plataformas obsoletas
print(obsolete_platforms_cumul_sales.head(5).reset_index())
```

```python
#¿Cuánto tardan generalmente las nuevas plataformas en aparecer y las antiguas en desaparecer?
video_games_with_year = video_games[video_games['year_of_release']>0]
first_active_year = video_games_with_year.groupby('platform')['year_of_release'].min()
last_active_year = video_games.groupby('platform')['year_of_release'].max()
life_cicle_platform = pd.merge(first_active_year,last_active_year,on=['platform'], how= 'left').reset_index()
life_cicle_platform.columns = ['platform','first_active_year','last_active_year']
life_cicle_platform = life_cicle_platform.sort_values(by='first_active_year', ascending=True)
life_cicle_platform['life_cicle_years'] = life_cicle_platform['last_active_year'] - life_cicle_platform['first_active_year']

print(life_cicle_platform)
print()
play_station = ['PS','PS2','PS3','PS4']
life_cicle_platform_playstation = life_cicle_platform.query("platform in @play_station").copy()

life_cicle_platform_playstation['years_since_previous_platform'] = (life_cicle_platform_playstation['first_active_year'] 
                                                                    - life_cicle_platform_playstation['first_active_year'].
                                                                    shift(1))


print(life_cicle_platform_playstation)
print()
print('El número de años promedio para que aparezca una nueva play sation es: ', 
      life_cicle_platform_playstation['years_since_previous_platform'].mean())
print()
print('El ciclo promedio de vida de una play station es: ', 
      life_cicle_platform_playstation['life_cicle_years'].median())
print()

xbox = ['XB','X360','XOne']
life_cicle_platform_xbox = life_cicle_platform.query("platform in @xbox").copy()

life_cicle_platform_xbox['years_since_previous_platform'] = (
    life_cicle_platform_xbox['first_active_year'] - life_cicle_platform_xbox['first_active_year'].shift(1)
)


print(life_cicle_platform_xbox)
print()
print('El número de años promedio para que aparezca una nueva xbox es: ', 
      life_cicle_platform_xbox['years_since_previous_platform'].mean())
print()
print('El ciclo promedio de vida de una xbox es: ', 
      life_cicle_platform_xbox['life_cicle_years'].median())


#platform_by_year = video_games_with_year.pivot_table(
#    index = 'year_of_release',
#    columns = 'platform',
#    values = 'total_sales',
#    aggfunc = 'sum'
#)

#print(platform_by_year)
```

```python
# Cómo las reseñas de usuarios y profesionales afectan las ventas de una plataforma popular. 
# Crear un gráfico de dispersión y calcular la correlación entre las reseñas y las ventas.

platform_ps4 = video_games_since_2008[video_games_since_2008['platform'] == 'PS4'].copy()

print(platform_ps4[['name','platform','year_of_release','genre','critic_score','user_score','user_score_numeric','total_sales']].head(20))

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(
    x=platform_ps4['critic_score'], 
    y=platform_ps4['total_sales'],      
    alpha=0.5        
)

plt.title('Critic score and sales correlation PS4')
plt.xlabel('Critic Score')
plt.ylabel('Total Sales')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

critic_score_sales_correlation = platform_ps4[['critic_score', 'total_sales']].corr().iloc[0, 1]
print(f"Critic score and sales correlation: {critic_score_sales_correlation:.2f}")
```

```python
# Excluir Critic Score & total sales iguales a cero
platform_ps4_critic = platform_ps4[(platform_ps4['critic_score'] > 0) & (platform_ps4['total_sales'] > 0)]

plt.figure(figsize=(10, 6))
plt.scatter(
    x = platform_ps4_critic['critic_score'], 
    y = platform_ps4_critic['total_sales'],      
    alpha = 0.5        
)

plt.title('Critic score and sales correlation PS4')
plt.xlabel('Critic Score')
plt.ylabel('Total Sales')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

critic_score_sales_correlation = platform_ps4_critic [ [ 'critic_score', 'total_sales' ] ].corr().iloc[0, 1]
print(f"Critic score and sales correlation: {critic_score_sales_correlation:.2f}")
```

```python
# Carcular p-valor > 0.05, para establecer si la correlación no es estadisticamente significativa. 
from scipy.stats import pearsonr
corr, p_value = pearsonr(platform_ps4_critic['critic_score'], platform_ps4_critic['total_sales'])
print(f"p-valor > 0.05 se considera una correlación estadísticamente significativa:: {p_value:.4f}") 
```

## Prueba de hipótesis

```python
# Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

# H0: μ_user_score_Xbox = μ_user_score_PC (no hay diferencia en ingresos mensuales)
# H1: μ_user_score_Xbox ≠ μ_user_score_PC (hay diferencia)


alpha = 0.05

xone_user_score = video_games_since_2008.query('platform == "XOne" & user_score_numeric > 0')['user_score_numeric']
pc_user_score = video_games_since_2008.query('platform == "PC" & user_score_numeric > 0')['user_score_numeric']


results = st.ttest_ind(xone_user_score,pc_user_score ,equal_var=False)

print('valor p: ', results.pvalue)

if results.pvalue < alpha: 
    print("Rechazamos H0: hay diferencia significativa en la calificación promedio.")
else:
    print("No rechazamos H0: no hay evidencia de diferencia.")
```

```python
# Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.

alpha = 0.05

action_user_score = video_games_since_2008.query('genre == "Action" & user_score_numeric > 0')['user_score_numeric']
sports_user_score = video_games_since_2008.query('genre == "Sports" & user_score_numeric > 0')['user_score_numeric']


results = st.ttest_ind(action_user_score,sports_user_score ,equal_var=False)

print('valor p: ', results.pvalue)

if results.pvalue < alpha: 
    print("Rechazamos H0: hay diferencia significativa en la calificación promedio.")
else:
    print("No rechazamos H0: no hay evidencia de diferencia.")
```

<div class="alert alert-block alert-info">
<b>Comentario del estudiante</b> <a class="tocSkip"></a>

- Se formula la hipótesis nula se formula a partir de lo que queremos refutar con las dos pruebas, que no hay diferencia significativa entre las calificaciones promedio. 
- Utilizamos un umbral del 5% como medida común de significación estadística. Un α del 5% implica aceptar que hay un 5% de probabilidad de concluir erróneamente que existe un efecto cuando no lo hay.
